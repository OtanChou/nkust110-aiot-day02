```python
url = 'https://www.nkust.edu.tw/p/403-1000-1363-{}.php' 
pages = [url.format(page) for page in range(1, 49)]
print(pages)
```

```python
import requests, time, json
from bs4 import BeautifulSoup
all_news_link = list()

# for url in pages:
for url in pages[:3]: #小範圍測試
    try:
        html = requests.get(url).text  #用get的方式，以文字檔紀錄抓取資料
        soup = BeautifulSoup(html, 'html.parser')   #soup解析網頁內容
        links = soup.find_all(class_ = "mtitle") 
        # print(links)
        for link in links:
            #透過新聞連結，以利後續抓取新聞內容
            item = dict()
            item['title'] = link.a.text.strip()  #抓取新聞標題
            item['href'] = link.a['href']  #抓取新聞連結
            item['date'] = link.i.text  #抓取新聞時間
            all_news_link.append(item)
#             print(item)
    except:
        pass
    time.sleep(3)
    

with open("nkustnews.json", "w", encoding = "utf-8") as fp:
    json.dump(all_news_link, fp)
print(all_news_link)
```
